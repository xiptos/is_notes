```python
import gymnasium as gym
from stable_baselines3 import SAC
```

```python
# 1. Create the environment
env = gym.make("Pusher-v5", render_mode="rgb_array")
```

```python

# 2. Create a model
model = SAC("MlpPolicy", env, verbose=1, tensorboard_log="tb/")

# 3. Train the agent
model.learn(total_timesteps=80_000)
```

```python
model.save("pusher_sb3_agent")
```

```python
del model
```

```python
model = SAC.load("pusher_sb3_agent", env=env)

env = model.get_env()

#4. Evaluate or run a demo
obs = env.reset()
for _ in range(1000):
    action, _ = model.predict(obs, deterministic=True)
    obs, reward, done, info = env.step(action)
    env.render("human")
    if done:
        obs = env.reset()

env.close()
```
